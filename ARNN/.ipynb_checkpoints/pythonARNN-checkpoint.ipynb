# ARNN and vARNN

```python
import numpy as np
import copy
import matplotlib.pyplot as plt
%matplotlib inline
## for jupyter notebook or Colab
#%matplotlib notebook 
## for jupyter lab
#%matplotlib widget 
```

```python
def rnn_cell_forward(xt, a_prev, parameters):
    """
    Implements a forward step of the RNN-cell 

    Arguments:
    xt -- input at "t", numpy array of shape (1,2). 
    a_prev -- Hidden state at timestep "t-1", numpy array of shape (n_a, 2)
    parameters -- python dictionary containing:
                        W -- Weight matrix multiplying the input, numpy array of shape (1, n_a)
                        b --  Bias, numpy array of shape (n_a, 1)
                        sig -- private noise
                        alpha -- noise associated with reward r=1
    Returns:
    a_next -- next hidden state, of shape (n_a, 2)
    yt_pred -- prediction at timestep "t", numpy array of shape (n_y,1)
    cache -- tuple of values needed for the backward pass, contains (a_next, a_next_, abr, e, parameters)
    """
    # Retrieve parameters 
    W = parameters["W"] # 1, n_a
    b = parameters["b"] # 1
    if parameters['USE_VAR']:  # USE_VAR=0: vanilla, ARNN 1: use variation vARNN
        alpha = parameters["alpha"] # r = 1
        beta = parameters["beta"] # r = 0
    else:
        alpha = beta = 1.0
        
    
    _, n_a = W.shape
    #print(n_a, a_prev.shape)
    assert a_prev.shape == (n_a, 2), "Error!, Dimension does not match!"
    # compute cell
    # a_next:(n_a,2),  xt:[r,x], latest last,  accoridng to W
    a_next = np.concatenate((a_prev[1::,:], xt), axis=0)  
    ar_next = np.row_stack(a_next[:,0]) # r
    ax_next = np.row_stack(a_next[:,1]) # x
    e = np.random.randn(n_a,1) # save for the reparameterization trick
    abr = np.sqrt(alpha**2 *ar_next + beta**2 *(1-ar_next))
    a_next_ = (1.0 + abr*e)*ax_next
    yt_pred = W @ a_next_ + b #  try ReLu later
    
    # store values you need for backward propagation in cache
    cache = (a_next, a_next_, abr, e, parameters)
    
    return a_next, yt_pred, cache

def relu(x): 
    """
    Elementwise ReLu function
        x: numpy array
        return : np array of same shape
    """
    return max(0.0, x)

def rnn_forward(X, a0, parameters):
    """
    Implement the forward propagation of the recurrent neural network

    Arguments:
    X -- Input data for every time-step, of shape  (T, n_x).
    a0 -- Initial hidden state, of shape (n_a, n_x)
    parameters -- python dictionary 

    Returns:
    a -- Hidden states for every time-step, numpy array of shape (n_a, T_x)
    y_pred -- Predictions for every time-step, numpy array of shape (n_y, T_y)
    caches -- tuple of values needed for the backward pass, contains list of cache for each step
    """
    caches = []
    T, n_x = X.shape # (T,2)
    n_y, n_a = parameters['W'].shape #(1,20)
    
    a = np.zeros((n_a,n_x,T)) #(20,2,T)
    y_pred = np.zeros((n_y, T))# (1,T)
    
    a_next = a0 # initialize the cell state
    
    for t in range(T):
        a_next, yt_pred, cache = rnn_cell_forward(np.column_stack(X[t,:]), a_next, parameters)
        a[:,:,t] = a_next
        y_pred[:,t] = yt_pred
        caches.append(cache)
    
    return a, y_pred, caches
```

```python
def rnn_cell_backward(dy_pred, cache):
    """
    Implements the backward pass for single time-step through RNN cell.

    Arguments:
    dy_pred --dL/dy_pred, derivative wrt y_pred for this time step
    cache -- python dictionary containing useful values (output of rnn_cell_forward())

    Returns:
    gradients
                        
    """
    # Retrieve values from cache
    (a_next, a_next_, abr, e, parameters) = cache
    
    # Retrieve parameters from "parameters"
    W = parameters["W"] # 1, n_a
    b = parameters["b"] 
    
    # compute the gradients
    if parameters['USE_VAR']:
        alpha = parameters["alpha"] 
        beta = parameters["beta"] 
        dW = dy_pred*a_next_.T #a_next_:(n_a,1) latest last
        db = dy_pred*1.0 #dy_pred.shape: (1,1)
        ar_next = np.row_stack(a_next[:,0])
        ax_next = np.row_stack(a_next[:,1])
        dalpha = dy_pred*(W @ ((1.0/abr)*ar_next*e*ax_next))*alpha
        dbeta = dy_pred*(W @ ((1.0/abr)*(1-ar_next)*e*ax_next))*beta
        gradients = {"dW": dW, "db": db, "dalpha": dalpha, "dbeta": dbeta}
    else:
        dW = dy_pred*a_next_.T
        db = dy_pred*1.0
        gradients = {"dW": dW, "db": db}

    return gradients

def rnn_backward(dy, caches):
    """
    Implement the backward pass of RNN through an entire sequence of input data.

    Arguments:
    dy -- dL/dy_pred on all steps
    caches -- tuple containing information from the forward pass (rnn_forward)
    Returns:
    gradients -- aggregated gradients
    """
    (a1, a1_, abr1, e1, parameters) = caches[0]
    
    # Retrieve dimensions
    n_y, T = dy.shape
    n_a, n_x = a1.shape
    # initialize the gradients with the right sizes 
    dW = np.zeros((n_y,n_a))
    db = np.zeros((n_y,1)) 
    dalpha = np.zeros((1,1)) 
    dbeta = np.zeros((1,1)) 
    
    # backward pass through all the time steps and cumulate gradients
    for t in reversed(range(T)):
        gradients = rnn_cell_backward(dy[:,t], caches[t])
        if parameters['USE_VAR']:
            dW += gradients["dW"] 
            db += gradients["db"] 
            dalpha += gradients["dalpha"]
            dbeta += gradients["dbeta"]
        else:
            dW += gradients["dW"] 
            db += gradients["db"] 
    # cache the gradients
    if parameters['USE_VAR']:
        gradients = {"dW": dW, "db": db, "dalpha": dalpha, "dbeta": dbeta}
    else:
        gradients = {"dW": dW, "db": db}
    
    return gradients
```

```python
def initialize_parameters(parameters):
    ## random sample around set values to initialize paras
    np.random.seed(1) # for reproducibility
    n_y, n_a = parameters['W'].shape
    parameters['W'] *= np.abs(1 + 0.2*np.random.randn(n_y, n_a))
    parameters['b'] *= (1 + 0.2*np.random.randn(n_y, 1))
    if  parameters['USE_VAR']:
        parameters['alpha'] *= (1 + 0.2*np.random.randn(1, 1))
        parameters['beta'] *= (1 + 0.2*np.random.randn(1, 1))
        
    return 

def update_parameters(parameters, gradients, lr):
    ## W = W - lr*dW
    parameters['W'] -= lr * gradients['dW'] 
    parameters['b'] -= lr * gradients['db']
    if  parameters['USE_VAR']:
        parameters['alpha'] -= lr * gradients['dalpha']
        parameters['beta'] -= lr * gradients['dbeta']
    
    return parameters

def loss_ARNN(y, y_pred): 
    """
    Vanilla MSE loss and dLoss/dy_pred
    """
    assert len(y) == len(y_pred), "Error! y, y_pred size does not match!"
    loss = np.mean((y_pred - y)**2, axis=1) 
    dy = np.asarray(2*(y_pred - y)) # dL / dy_pred
    
    return loss, dy

def loss_vARNN(y, y_pred, X, parameters, LossMethod = 'MSE', SparseReg = True):
    """
    Loss function
    Arguments: y, y_pred: actual and predicted time series
               X: input consists of (r,x) reward and input time series
                Parameters: genearator model or learning model parameters
                LossMethod: 'MSE': standard MSE, 'rMSE': random weight MSE, 'wMSE': variance weighted MSE
                SparseReg: regularization for sparsity 
    Returns: Loss: total loss 
             dy: dLoss/dy_pred (1,T)
            
    """
    assert len(y) == len(y_pred), "Error! y, y_pred size does not match."
    r, x = X[:,0], X[:,1] #
    assert y.size == r.T.size, "Error! y and r do not match."
    W = parameters['W']
    if parameters['USE_VAR']:
        alpha = parameters['alpha']
        beta =  parameters['beta']
    else:
        alpha = beta = 1.0
   
    
    if LossMethod == 'wMSE':
        r_var = (alpha**2 *r + beta**2 *(1 - r)).T # variance driven by reward
        _, n_a = W.shape
        index = np.arange(n_a, len(r_var))
        weight = np.zeros(y_pred.T.shape)
        for i in range(len(index)):
            weight[index[i]] = 1/(W**2 @ r_var[i:index[i]])
            
        weight_norm = weight/np.mean(weight)
        loss = np.mean(weight_norm.T * (y_pred - y)**2, axis = 1 )
        dy = np.asarray(2 * weight_norm.T * (y_pred - y))
        
    elif LossMethod == 'rMSE':
        random_weight = np.random.uniform(0, 1, y.T.shape).transpose()
        random_weight = random_weight/np.mean(random_weight)
        loss = np.mean(random_weight*(y_pred - y)**2, axis = 1)
        dy = np.asarray(2*random_weight*(y_pred - y))
    
    else: # 'MSE' as deflaut
        loss = np.mean((y_pred - y)**2, axis=1) 
        dy = np.asarray(2*(y_pred - y)) # dL / dy_pred
    
    ## Regularization
    if SparseReg:
        sparsity = 0.5
        lambda_coeff = 0.01
        c1 = 3.0
        c2 = 3.0
        panelty_noise = (alpha**2/beta**2 + beta**2/alpha**2)
        panelty_coeff = np.mean(2/(1 + np.exp(-c1 * np.abs(W)**(1/c2)))) - 1.0
        panelty = lambda_coeff *( (1/(sparsity - 0.001) -1.0) * panelty_coeff + panelty_noise)
        loss = (1 + panelty) * loss
    
    return loss, dy


## clip the gradients, store in place
def clip(gradients, maxValue = 10.0):
    #is it in place? might need to check the data structure in gradients
    for gradient in gradients.values(): 
        np.clip(gradient, -maxValue, maxValue) 
    
    return 


### 
def optimize(X, y, parameters, learning_rate = 0.001, LossMethod = 'MSE', SparseReg = True):
    """
    one step optimization
    assemble everything together
    
    """
    a, y_pred, caches = rnn_forward(X, a0, parameters)
    
    #loss, dy = loss_ARNN(y, y_pred)
    loss, dy = loss_vARNN(y, y_pred, X, parameters, LossMethod, SparseReg)
    
    gradients = rnn_backward(dy, caches)
    
    clip(gradients)

    parameters = update_parameters(parameters, gradients, learning_rate)
    
    return loss, gradients
```

```python
# genearate a time series
def tp_generator(T, parameters):
    """
    T: length of the time series
    parameters: paras for generating the series
    
    """
    #np.random.seed(1)
    W = parameters['W']
    b = parameters['b']
    if  parameters['USE_VAR']:
        alpha = parameters['alpha']
        beta = parameters['beta']
    else:
        alpha = 0.0
        beta = 0.0
    k_reward= parameters['k_reward']
    sig = parameters['sig']
    tp = np.zeros((T,1))
    r = np.zeros((T,1))
    n_y,n_a = W.shape
    n_x = 2
    if 1:
        # less efficient way
        a_prev = np.zeros((n_a,n_x))
        xt = np.zeros((1,n_x))
        for t in range(T):
            a_next, yt_pred, _ = rnn_cell_forward(xt, a_prev, parameters)
            tp[t,0] = yt_pred + sig*np.random.randn(n_y,1)
            r[t,0] = (np.abs(tp[t]) < k_reward).astype(float)
            xt = np.column_stack((r[t,0],tp[t,0]))#np.concatenate((r[t],tp[t]), axis = 1)
            a_prev = a_next
    else:
        for i in range(T):
            if i < n_a:  # initalize with random
                tp[i] = b + sig*np.random.randn(n_y, 1)
            else:
                er = np.random.randn(n_a,1)
                abr = np.sqrt(alpha**2 *r[i-n_a:i] + beta**2 *(1-r[i-n_a:i]))
                tp[i] = W @ ((1.0 + abr*er) * tp[i-n_a:i]) + b + sig*np.random.randn(1)

        r = (np.abs(tp) < k_reward).astype(float)

    return tp, r
```

```python
# simulate a forward pass
# optimize the parameters with simulated data
T = 500
n_a = 20
x = np.column_stack(np.arange(n_a -1,-1,-1))
W = 0.2*np.exp(-x/3) # true W
b = 0
alpha = 0.1
beta = 0.3
sig = 0.1
k_reward = 0.1
n_y,n_a = W.shape
parameters_= {}
parameters_['W'] = copy.deepcopy(W)
parameters_['b'] = b
parameters_['alpha'] = alpha
parameters_['beta'] = beta
parameters_['k_reward'] =k_reward
parameters_['sig'] = sig
parameters_['USE_VAR'] = True

tp, r = tp_generator(T, parameters_)
X = np.concatenate((r, tp), axis = 1)
y = np.concatenate((tp[1::,:],np.zeros((1,1))), axis = 0).T

a0 = np.zeros((n_a,2))
a, y_pred, _ = rnn_forward(X, a0, parameters_) # one step forward prediction with correct paras
trial = np.arange(T)

# test variance weighted
r = X[:,0]
alpha = parameters_['alpha']
beta =  parameters_['beta']
r_var = (alpha**2*r + beta**2*(1 - r)).T # variance driven by reward
W = parameters_['W']
_, n_a = W.shape
index = np.arange(n_a, len(r_var))
weight = np.zeros(y_pred.T.shape)
for i in range(len(index)):
    weight[index[i]] = 1/( W**2 @ r_var[i:index[i]] )

weight_norm = weight/np.mean(weight)
loss = np.mean(weight_norm.T * (y_pred - y)**2, axis = 1 )
dy = np.asarray(2 * weight_norm.T * (y_pred - y))
print('Loss = ', loss)

##
%matplotlib inline
figsz = (15,10)
fontsz = 15
fig = plt.figure(figsize=figsz )
ax1 = fig.add_subplot(311)
ax1.plot(trial, np.zeros(trial.shape), 'g:')
ax1.plot(trial, tp, 'k.')
ax1.plot(trial[:-1], y_pred[:,1::].T, 'r-', linewidth=1)
ax1.set_ylabel("y, y_pred", fontsize = fontsz)
ax = fig.add_subplot(312, sharex=ax1)
ax.scatter(index,weight_norm[index], s = 1, c = 'k')
ax.set_ylabel("Weight for Loss", fontsize = fontsz)
ax = fig.add_subplot(313, sharex=ax1)
ax.scatter(index,r[index], s = 1, c = 'k')
ax.set_xlabel("Trial #", fontsize = fontsz), ax.set_ylabel("Reward", fontsize = fontsz);
```

# GRU with Tensorflow

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import datetime
%load_ext tensorboard
```

```python
def sliding_windows(data, seq_length):
    x = []
    y = []
    n, p = data.shape 
    for i in range(n-seq_length-1):
        _x = data[i:i+seq_length,:].T
        _y = data[i+seq_length, 1]
        x.append(_x[:,::-1]) # reversed!
        y.append(_y)

    return np.array(x),np.array(y)

n_a = 20 # dim of cell/state units
_, n_x = X.shape
X_, y_ = sliding_windows(X, n_a)
N = len(y_)
split = 0.8
sample_ind = np.random.permutation(N)
train_sample = sample_ind[:int(N*split)]
test_sample = sample_ind[int(N*split):]
X_train, y_train = X_[train_sample,:,:], y_[train_sample]
X_test, y_test = X_[test_sample,:,:], y_[test_sample]
```

```python
model = keras.Sequential()
model.add(layers.GRU(n_y,activation='relu',recurrent_activation='sigmoid', input_shape=(n_x, n_a))) # GRU recurrent
model.add(layers.Dense(n_y)) # output
print(model.summary())
```

```python
model.compile(
    optimizer='sgd',
    loss = 'mean_squared_error',
    metrics=['MeanSquaredError']
)
log_dir = "tf_logs/GRU/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
fit_history = model.fit(X_train, y_train, 
                        validation_data = (X_test, y_test), 
                        batch_size= 16, epochs = niter, 
                       callbacks = [tensorboard_callback])
```
